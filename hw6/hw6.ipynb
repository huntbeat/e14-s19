```python
%matplotlib inline
from scipy.stats import binom
from scipy.stats import poisson
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### 8.16

```python
x = [i for i in range(0,11)]

binom0 = binom.pmf(x, n=10, p=0.05)
pois0 = poisson.pmf(x, mu=0.5)

binom1 = binom.pmf(x, n=20, p=0.5)
pois1 = poisson.pmf(x, mu=10)

plt.figure(1)
plt.plot(x, binom0)
plt.title('Problem A')
plt.plot(x, pois0)
plt.legend(['Binomial', 'Poisson'])
plt.xlabel('X = x')
plt.ylabel('Probability')

plt.figure(2)
plt.plot(x, binom1)
plt.title('Problem B')
plt.plot(x, pois1)
plt.legend(['Binomial', 'Poisson'])
plt.xlabel('X = x')
plt.ylabel('Probability')
plt.show()
```

The Binomial and Poisson distributions in the first plot match nicely, while the two in the second plot deviate from each other.

### 8.26

```python
frame = pd.read_csv('prob8.26.csv',header=0,delimiter=",")
avg = np.mean(frame['Available Pumps'])
print ("mean = %4.2f" % avg)

bin_array = np.arange(0,13,1)   # Generate bins 8 for process A
frame['Available Pumps'].hist(bins=bin_array, density=True)

p_of_x = binom.pmf(bin_array, n=10, p=avg/10)
plt.plot(bin_array, p_of_x, 'bo', ms=8, label='binom pmf')
plt.vlines(x, 0, p_of_x, colors='b', lw=5, alpha=0.5)
plt.xlabel("Available Pumps",fontsize=20)
plt.ylabel("Frequency",fontsize=20)

plt.show()
```

Knowing that the pumps are identical and assuming that their performances are independent, we can use a Binomial distribution with n equal to the total number of pumps and p equal to the average number of available pumps divided by the total. We can see that the Binomial distribution does a good job of approximating the actual distribution.
