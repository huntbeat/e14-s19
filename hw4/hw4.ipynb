## Problem 1. std. dev. = 20

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
n = 10000
mu_x = 150.0
sigma_x = 20.0
x_sample = np.random.normal(mu_x,sigma_x,n)
bins_x = np.arange(50,250,5)                # NOTE: You should change the range and bin width as you change sigma_x
plt.hist(x_sample,density=True,bins=bins_x)
plt.title('Normal Distribution with Mean=150 and Standard Deviation=20')
plt.xlabel("Values of X")
plt.ylabel("Probability Density")
```

## std. dev. = 50

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
n = 10000
mu_x = 150.0
sigma_x = 50.0
x_sample = np.random.normal(mu_x,sigma_x,n)
bins_x = np.arange(0,300,10)                # NOTE: You should change the range and bin width as you change sigma_x
plt.hist(x_sample,density=True,bins=bins_x)
plt.title('Normal Distribution with Mean=150 and Standard Deviation=20')
plt.xlabel("Values of X")
plt.ylabel("Probability Density")
```

You can notice that the range of the distribution get smaller when the standard deviation changes from 20 to 5, and the bins need to be smaller for the distribution to look like a bell curve, or a higher resolution. The opposite happens when it moves from 20 to 50 -- the range becomes much wider, and bigger bins help make curve look like a bell.
