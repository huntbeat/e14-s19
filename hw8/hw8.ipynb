```python
prob21 = pd.read_csv("prob12.21.csv")

print(mean(prob21))
print(median(prob21))
print(var(prob21))
prob21 = prob21.T
title("Problem 21")
hist(prob21)

print(sum(prob21.T > 2))
print(len(prob21.T))

print(mean(prob21.T[0:9]))
print(mean(prob21.T[10:19]))
print(mean(prob21.T[20:29]))

show()
```

#### Problem 21:

a. The mean is 1.988, the median is 1.360, and the variance is 3.333. The shape is not surprising because we expect time between violations to be shorter than longer.

b. A Exponential distribution. The theoretical probability of going more than 2 months without a safety violation is 0.37. This is similar to the sample data, where 10 out of 30 data points exceed 2 months.

c. The mean of time between violations for Operator A is shorter than that of B, but is longer than that of C. Therefore, the data partially supports the suggestion that Operator A is relatively more safety-conscious.
